```python
X = df.drop(['Wins'],axis=1)
Y = df['Wins']
```

```python
from sklearn.decomposition import PCA
pca = PCA()
x_p = pca.fit_transform(X_scale)
plt.figure(figsize=(14,14))
plt.plot(np.cumsum(pca.explained_variance_ratio_), 'ro-')
plt.xlabel('Number of Components')
plt.ylabel('Variance ')
plt.grid()
```

```python
pca_new = PCA(n_components=4)
x_new = pca_new.fit_transform(X_scale)
```

```python
principle_x=pd.DataFrame(x_new,columns=np.arange(4))
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif= pd.DataFrame()
vif['Features']= principle_x.columns
vif['VIF']= [variance_inflation_factor(principle_x.values,i) for i in range(principle_x.shape[1])]

vif
```

# building model

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.22,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score,mean_squared_error
```

```python
def train_n_ret_scores(model):
    model.fit(X_train,y_train)
    
    y_preds = model.predict(X_test)
    print('MSE:',mean_squared_error(y_test,y_preds))
    print('R2_Score',r2_score(y_test,y_preds))
    score = cross_val_score(model, X, Y, cv =3)
    print("Mean CV Score :",score.mean())
```

```python
print('linear regression')
train_n_ret_scores(LinearRegression())
print('==============================================================================================================')
print('ridge regression')
train_n_ret_scores(Ridge(alpha=0.001))
print('==============================================================================================================')
print('lasso regression')
train_n_ret_scores(Lasso(alpha=0.00003))
print('==============================================================================================================')
print('elasticnet regression')
train_n_ret_scores(ElasticNet(alpha=0.0001))
print('==============================================================================================================')
print('svr regression')
train_n_ret_scores(SVR(kernel='linear',C=10,gamma=0.001))
print('==============================================================================================================')
print('decisiontree regression')
train_n_ret_scores(DecisionTreeRegressor(criterion='mse'))
print('==============================================================================================================')
print('randomforest regression')
train_n_ret_scores(RandomForestRegressor(n_estimators=500))
print('==============================================================================================================')
print('gradientboosting regression')
train_n_ret_scores(GradientBoostingRegressor())
print('==============================================================================================================')
print('adaboost regression')
train_n_ret_scores(AdaBoostRegressor(base_estimator=LinearRegression()))

```

```python
Final = AdaBoostRegressor(learning_rate=0.0001, n_estimators=100)
Final.fit(X_train,y_train)
y_pred=Final.predict(X_test)

print('\033[1m'+' R2 Score of Final Model :'+'\033[0m')
print(r2_score(y_test,y_pred)) 
print('\n')
```
